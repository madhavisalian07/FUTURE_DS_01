```python
import pandas as pd

data = {
    "Region": ["East", "West", "South", "North"],
    "Sales": [1000, 2000, 1500, 3000]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
import matplotlib.pyplot as plt

df.plot(x="Region", y="Sales", kind="bar")

plt.title("Sales by Region")

plt.show()
```
