```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Region": ["East", "West", "South", "North"],
    "Sales": [1000, 2000, 1500, 3000]
}

df = pd.DataFrame(data)

print(df.head())

# Bar chart
df.plot(x="Region", y="Sales", kind="bar")
plt.title("Sales by Region")
plt.show()

# Pie chart (correct way)
plt.figure()
plt.pie(df["Sales"], labels=df["Region"], autopct="%1.1f%%")
plt.title("Sales Distribution by Region")
plt.show()
```
